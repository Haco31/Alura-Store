### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
precio_tienda = tienda["Precio"]
precio_tienda2 = tienda2["Precio"]
precio_tienda3 = tienda3["Precio"]
precio_tienda4 = tienda4["Precio"]
```

```python
suma_tienda1 = precio_tienda.sum()
suma_tienda2 = precio_tienda2.sum()
suma_tienda3 = precio_tienda3.sum()
suma_tienda4 = precio_tienda4.sum()
```

```python
print(f"EL ingreso total de la tienda 1 es: {suma_tienda1}")
print(f"El ingreso total de la tienda 2 es: {suma_tienda2}")
print(f"El ingreso total de la tienda 3 es: {suma_tienda3}")
print(f"El ingreso total de la tienda 4 es: {suma_tienda4}")
```

```python
tienda = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [suma_tienda1, suma_tienda2, suma_tienda3, suma_tienda4]
colores = ["red", "blue", "green", "orange"]

plt.bar(tienda, ingresos, color=colores)
plt.title("Ingreso por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")
plt.show()
```

# 2. Ventas por categoría

```python
categoria = tienda["Categoría del Producto"]
categoria2 = tienda2["Categoría del Producto"]
categoria3 = tienda3["Categoría del Producto"]
categoria4 = tienda4["Categoría del Producto"]
```

```python
categorias_unicas = categoria.unique()
categorias_unicas2 = categoria2.unique()
categorias_unicas3 = categoria3.unique()
categorias_unicas4 = categoria4.unique()
```

```python
print(f"En la Tienda 1 existen las categorías: {categorias_unicas}")
print(f"En la Tienda 2 existen las categorías: {categorias_unicas2}")
print(f"En la Tienda 3 existen las categorías: {categorias_unicas3}")
print(f"En la Tienda 4 existen las categorías: {categorias_unicas4}")
```

```python
conteo_categorias = categoria.value_counts()
conteo_categorias2 = categoria2.value_counts()
conteo_categorias3 = categoria3.value_counts()
conteo_categorias4 = categoria4.value_counts()
```

```python
print(f"{conteo_categorias}\n")
print(f"{conteo_categorias2}\n")
print(f"{conteo_categorias3}\n")
print(f"{conteo_categorias4}\n")
```

```python
print(f"Calificación por {calificacion_categoria1}\n")
print(f"Calificación por {calificacion_categoria2}\n")
print(f"Calificación por {calificacion_categoria3}\n")
print(f"Calificación por {calificacion_categoria4}\n")
```

```python
print(f"La calificación promedio de la tienda 1 es: {calificacion_tienda}")
print(f"La calificación promedio de la tienda 2 es: {calificacion_tienda2}")
print(f"La calificación promedio de la tienda 3 es: {calificacion_tienda3}")
print(f"La calificación promedio de la tienda 4 es: {calificacion_tienda4}")
```

```python
import numpy as np
tienda1 = (["Muebles", "Artículos para el Hogar"], [465, 171])
tienda2 = (["Muebles", "Artículos para el Hogar"], [442, 181])
tienda3 = (["Muebles", "Instrumentos musicales"], [499, 177])
tienda4 = (["Muebles", "Instrumentos mucisales"], [480, 170])

fig, axes = plt.subplots(2,2, figsize=(10, 8))

def plot_tienda_data(ax, data, tienda_num):
  categorias, cantidades = data
  ax.bar(categorias, cantidades)
  ax.set_title(f'Tienda {tienda_num}')
  ax.set_ylabel('Cantidad')

plot_tienda_data(axes[0,0], tienda1, 1)
plot_tienda_data(axes[0,1], tienda2, 2)
plot_tienda_data(axes[1,0], tienda3, 3)
plot_tienda_data(axes[1,1], tienda4, 4)

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
costo_envio_tienda = tienda["Costo de envío"].mean().round(2)
costo_envio_tienda2 = tienda2["Costo de envío"].mean().round(2)
costo_envio_tienda3 = tienda3["Costo de envío"].mean().round(2)
costo_envio_tienda4 = tienda4["Costo de envío"].mean().round(2)
```

```python
print(f"El costo de envío promedio de la tienda 1 es de: {costo_envio_tienda}\n")
print(f"El costo de envío promedio de la tienda 2 es de: {costo_envio_tienda2}\n")
print(f"El costo de envío promedio de la tienda 3 es de: {costo_envio_tienda3}\n")
print(f"El costo de envío promedio de la tienda 4 es de: {costo_envio_tienda4}\n")
```

```python
tienda = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
costo_envio = [costo_envio_tienda, costo_envio_tienda2, costo_envio_tienda3, costo_envio_tienda4]

plt.step(tienda, costo_envio)
plt.title("Ingreso por tienda")
plt.xlabel("Tiendas")
plt.ylabel("Ingresos")
plt.show()
```
